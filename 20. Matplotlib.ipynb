#### `Draw line using coordinate values`

````python
# Plot line from point(2, 3) to (3, 5)

# Create two numpy arrays: one with x-coordinate values 
# and other with y-coordinate values
xpoints = np.array([2, 5])
ypoints = np.array([3, 10])

# use plot() method in matplotlib module 
# plot takes two arguments: first-array containing x-coordinates
# second-array containing y-coordinates
plt.plot(xpoints, ypoints)

# show() method to visualize the plot done
plt.show()
````

````python
# Plot line from point(2, 3) to (3, 5) to (4, 6) to (5, 3)
xpoints = np.array([2, 3, 4, 5])
ypoints = np.array([3, 5, 6, 3])

plt.plot(xpoints, ypoints)
plt.show()
````

````python
# Plotting without line
# third argument is marker
plt.plot(xpoints, ypoints, 'o')
plt.show()
````

````python
# Plotting without line
# third argument is marker
# Highlight the points using 'marker' argument.
plt.plot(xpoints, ypoints, marker='o')
plt.show()
````

#### `Size of the marker with 'ms' argument`

````python
plt.plot(xpoints, ypoints, 's:g', ms=10)
plt.show()
````

#### `Color of edge of marker with 'mec' argument`

````python
plt.plot(xpoints, ypoints, 's:g', ms=10, mec='r')
plt.show()
````

#### `Color of face of marker with 'mfc' argument`

````python
plt.plot(xpoints, ypoints, 's:g', ms=10, mec='r', mfc='k')
plt.show()
````

#### `Hexadecimal values for colors`

````python
plt.plot(xpoints, ypoints, 's:g', ms=10, mec='r', mfc='#5CFA50')
plt.show()
````

#### `Control lines`
- style using 'ls' argument
- color using 'color' or 'c' argument
- line width using 'linewidth' argument 

````python
plt.plot(xpoints, ypoints, ls=':', c='g', linewidth='2')
plt.show()
````

#### `Plot multiple lines`

````python
xpoints_1, ypoints_1 = xpoints+1, ypoints+1
plt.plot(xpoints, ypoints)
plt.plot(xpoints_1, ypoints_1)
plt.show()
````

````python
xpoints_1, ypoints_1 = xpoints+1, ypoints+1

plt.plot(xpoints, ypoints, xpoints_1, ypoints_1)
plt.show()
````

#### `Labels and Title`

````python
plt.plot(xpoints, ypoints)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Title')
plt.show()
````

#### `Font Properties for Title and Labels`

````python
font1 = {'family':'serif','color':'blue','size':10}
font2 = {'family':'serif','color':'darkred','size':20}

plt.plot(xpoints, ypoints)
plt.xlabel('x-label', fontdict=font1)
plt.ylabel('y-label', fontdict=font1)
plt.title('Title', fontdict=font2)
plt.show()
````

#### `Position the Title`
- Place the title in 'left', 'right', and 'center' using `loc` attribute

````python
plt.plot(xpoints, ypoints)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Title', loc='center')
plt.show()
````

#### `Adding grid to the graph`

````python
plt.plot(xpoints, ypoints)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Title')
plt.grid()
plt.show()
````

#### `Choose which grid lines to display`

````python
plt.plot(xpoints, ypoints)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Title')
plt.grid(axis='y')
plt.show()
````

#### `line properties for grid`

### `Subplot`
- Multiple sub plots in single plot using `subplot()`
- takes 3 arguments:<br/>
    &emsp;first - number of rows<br/>
    &emsp;second - number of columns<br/>
    &emsp;third - plot number

````python
nrows = 1
ncols = 2

plt.subplot(nrows, ncols, 1)
x = np.array([1, 2, 3, 5])
y = np.array([3, 1, 9, 6])
plt.plot(x, y)

plt.subplot(nrows, ncols, 2)
x = np.array([2, 4, 8, 9])
y = np.array([0, 2, 1, 5])
plt.plot(x, y)
````

````python
nrows = 2
ncols = 2

plt.subplot(nrows, ncols, 1)
x = np.array([1, 2, 3, 5])
y = np.array([3, 1, 9, 6])
plt.plot(x, y)
plt.title("first")

plt.subplot(nrows, ncols, 2)
x = np.array([2, 4, 8, 9])
y = np.array([0, 2, 1, 5])
plt.plot(x, y)
plt.title("second")

plt.subplot(nrows, ncols, 3)
x = np.array([1, 2, 3, 5])
y = np.array([3, 1, 9, 6])
plt.plot(x, y)
plt.title("third")

plt.subplot(nrows, ncols, 4)
x = np.array([2, 4, 8, 9])
y = np.array([0, 2, 1, 5])
plt.plot(x, y)
plt.title("fourth")

plt.suptitle("Title")
````

#### `Scatter Plot`
- one dot for each observation
- requires two arrays of equal length: <br/>
&emsp; one for x-axis<br/>
&emsp; other for y-axis

````python
# scatter plot 
x = np.array([1, 2, 3, 4, 5, 6, 6, 7, 8, 9])
y = np.array([10, 2, 4, 5, 2, 8, 4, 1, 6, 11])

plt.scatter(x, y)
````

````python
# data for scatter plot
x = np.array([1, 2, 3, 4, 5, 6, 6, 7, 8, 9])
y = np.array([10, 2, 4, 5, 2, 8, 4, 1, 6, 11])
x1 = x + 1
y1 = y - 1
````

````python
# draw two plots in same figure
plt.scatter(x, y)
plt.scatter(x1, y1)
````

````python
# Adding custom colours to the plot
plt.scatter(x, y, color='g')
plt.scatter(x1, y1, color='k')
````

````python
# adding label to plot
# Adding custom colours to the plot
plt.scatter(x, y, color='g', label='green color')
plt.scatter(x1, y1, color='k', label="black color")
plt.legend()
````

````python
# Bar chart
x = np.array(["car", "bike", 'truck', 'bicycle'])
y = np.array([3, 8, 4, 6])

plt.bar(x,y)

# change color of bars 
# plt.bar(x,y,  color='g')

# width of the bar
# plt.bar(x, y, width=0.5)
````

````python
# Horizontal Bar chart
x = np.array(["car", "bike", 'truck', 'bicycle'])
y = np.array([3, 8, 4, 6])

plt.barh(x,y)

# height of bars 
plt.barh(x, y, height=0.2)
````

#### `Pie-chart`

````python
x = np.array([35, 25, 25, 15])

plt.pie(x)
````

Here, each portion is calculated using formula: `element/sum_of_all_elements`

````python
# Add labels to each portion of pie-chart
x = np.array([35, 25, 25, 15])
labels = ['car', 'bike', 'truck', 'bicycle']

plt.pie(x, labels=labels)
# startangle
````

````python
# explode the portion of pie-chart
x = np.array([35, 25, 25, 15])
labels = ['car', 'bike', 'truck', 'bicycle']
explodes = [0.1, 0, 0, 0]

plt.pie(x, labels=labels, explode=explodes)
# legend
````
